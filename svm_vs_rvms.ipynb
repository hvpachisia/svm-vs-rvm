### EDA

Attribute Information:

1) ID number 2) Diagnosis (M = malignant, B = benign) 3-32)

Ten real-valued features are computed for each cell nucleus:

a) radius (mean of distances from center to points on the perimeter) b) texture (standard deviation of gray-scale values) c) perimeter d) area e) smoothness (local variation in radius lengths) f) compactness (perimeter^2 / area - 1.0) g) concavity (severity of concave portions of the contour) h) concave points (number of concave portions of the contour) i) symmetry j) fractal dimension ("coastline approximation" - 1)

The mean, standard error and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. For instance, field 3 is Mean Radius, field 13 is Radius SE, field 23 is Worst Radius.

All feature values are recoded with four significant digits.

Missing attribute values: none

Class distribution: 357 benign, 212 malignant

```python
x=cancer[["radius_mean","texture_mean","smoothness_mean","compactness_mean","concavity_mean"]]
y=cancer["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
#randomly chose 0.30
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn import svm
from sklearn.metrics import jaccard_score
clf=svm.SVC(kernel='rbf',gamma="scale")
clf.fit(x_train,y_train)
predictions=clf.predict(x_test)

# 'M' is the label consider as the positive class
score = jaccard_score(y_test, predictions, pos_label='M')
print(score * 100)
```

Here we can visualize the SVM boundaries using two features.

```python
def plot_decision_boundaries(x_train, y_train, features):
    # Select the features for visualization
    x_visualize = x_train[features]

    # Ensure the data is numeric and handle NaN values
    x_visualize = x_visualize.apply(pd.to_numeric, errors='coerce')
    x_visualize = x_visualize.dropna()

    # Fit the SVM model
    clf = svm.SVC(gamma="scale")
    clf.fit(x_visualize, y_train)

    # Create a mesh to plot in
    h = .02  # step size in the mesh
    x_min, x_max = x_visualize.iloc[:, 0].min(), x_visualize.iloc[:, 0].max()
    y_min, y_max = x_visualize.iloc[:, 1].min(), x_visualize.iloc[:, 1].max()

    # Ensure the ranges are not too small
    x_range = np.arange(x_min, x_max, h)
    y_range = np.arange(y_min, y_max, h)

    if len(x_range) <= 1 or len(y_range) <= 1:
        print("Error: Meshgrid ranges are too small. Try adjusting the step size (h).")
        return

    xx, yy = np.meshgrid(x_range, y_range)

    # Plot decision boundaries
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])

    # Plot the decision boundaries
    Z_numeric = np.where(Z > 0, 1, 0)
    Z_numeric = Z_numeric.reshape(xx.shape)
    plt.contourf(xx, yy, Z_numeric, cmap=plt.cm.coolwarm, alpha=0.3)

    # Plot the data points with correct colors and labels
    colors = np.where(y_train == 'B', 'blue', 'red')
    plt.scatter(x_visualize.iloc[:, 0], x_visualize.iloc[:, 1], c=colors, cmap=plt.cm.coolwarm, edgecolors='k', marker='o', label='Benign (blue) vs. Malignant (red)')

    # Plot the support vectors
    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], facecolors='none', edgecolors='k', s=100, label='Support Vectors')

    # Plot decision boundary lines
    plt.contour(xx, yy, Z.reshape(xx.shape), colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

    # Set plot labels and title
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.title(f'SVM Decision Boundaries using Radial Kernel- {features[0]} and {features[1]}')

    # Add legend
    plt.legend()

    # Show the plot
    plt.show()


# Plot decision boundaries for two select features
plot_decision_boundaries(x_train, y_train, ["radius_mean", "texture_mean"])

plot_decision_boundaries(x_train, y_train, ["compactness_mean","concavity_mean"])
```

When it comes to an RVM, visualizing decision boundaries can be different compared to SVM because RVMs are probabilistic models. Instead of strict decision boundaries, we will visualize the probability contours.

```python
from sklearn.preprocessing import LabelEncoder

def plot_rvm_decision_boundaries(x_train, y_train, features):
    # Select the features for visualization
    x_visualize = x_train[features]

    # Ensure the data is numeric and handle NaN values
    x_visualize = x_visualize.apply(pd.to_numeric, errors='coerce')
    x_visualize = x_visualize.dropna()

    # Convert labels to numeric values
    label_encoder = LabelEncoder()
    y_numeric = label_encoder.fit_transform(y_train)

    # Fit the RVM model
    rvc_clf = RVC()
    rvc_clf.fit(x_visualize, y_numeric)

    # Create a mesh to plot in
    h = .02  # step size in the mesh
    x_min, x_max = x_visualize.iloc[:, 0].min(), x_visualize.iloc[:, 0].max()
    y_min, y_max = x_visualize.iloc[:, 1].min(), x_visualize.iloc[:, 1].max()

    # Ensure the ranges are not too small
    x_range = np.arange(x_min, x_max, h)
    y_range = np.arange(y_min, y_max, h)

    if len(x_range) <= 1 or len(y_range) <= 1:
        print("Error: Meshgrid ranges are too small. Try adjusting h.")
        return

    xx, yy = np.meshgrid(x_range, y_range)

    # Get probabilities from the RVM model
    probs = rvc_clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = probs[:, 1]  # Probability of positive class

    # Reshape for contour plot
    Z = Z.reshape(xx.shape)

    # Plot probability contours with labels
    contour = plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.3)
    plt.colorbar(contour, label='Probability of Malignant (Class 1)')

    # Plot the data points with correct colors and labels
    colors = np.where(y_numeric == 0, 'blue', 'red')
    plt.scatter(x_visualize.iloc[:, 0], x_visualize.iloc[:, 1], c=colors, cmap=plt.cm.coolwarm, edgecolors='k', marker='o', label='Benign (blue) vs. Malignant (red)')

    # Set plot labels and title
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.title(f'RVM Probability Contours - {features[0]} and {features[1]}')

    # Add legend
    plt.legend()

    # Show the plot
    plt.show()
# Example usage:

# Plot RVM probability contours for the selected features
plot_rvm_decision_boundaries(x_train, y_train, ["radius_mean", "texture_mean"])
```

The probability contours from the RVM model represent regions of the input space where the model assigns different probabilities to different classes. Each point in the input space is associated with a probability of belonging to a particular class.

In the case of benign vs. malignant tumors, the contours represent the decision boundaries between the two classes. The contours indicate the regions where the RVM model assigns a higher probability to one class compared to the other. The specific values on the contours represent the probability of belonging to the positive class. This demonstrates a more nuanced view of the model's predictions compared to a strict decision boundary.
